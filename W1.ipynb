```python
fu1 = cv2.imread("fusion1.jpeg")
fu2 = cv2.imread("fusion2.jpeg")

sh = fu1.shape
row = sh[0]
col = sh[1]
fu3 = np.uint8(np.zeros((sh[0],sh[1],3)))

for i in range(row):
    for j in range(col):
        for k in range(3):
            if(fu1[i][j][k] < fu2[i][j][k]):
                fu3[i][j][k] = fu1[i][j][k]
            elif(fu1[i][j][k] > fu2[i][j][k]):
                fu3[i][j][k] = fu2[i][j][k]
            else:
                fu3[i][j][k] = fu1[i][j][k]
cv2.imwrite("fusion1.png",fu1)
cv2.imwrite("fusion2.png",fu2)

cv2.imwrite("fusion4.png",fu3)
plt.imshow(fu3)
```
